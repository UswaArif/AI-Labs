Reproduce this figure.  **(4 marks)**

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(0,4*np.pi-1,0.1)  
sin_curve = np.sin(x)
cos_curve = np.cos(x)

# add code here
plt.plot(x,sin_curve,"blue", label='sin(x)')
plt.plot(x,cos_curve,"g", label='cos(x)')
plt.xlabel("x values from 0 to 4pi")
plt.ylabel('sin(x) and cos(x)')
plt.legend()
plt.show()
```

Reproduce this figure.  

```python
import numpy as np
import matplotlib.pyplot as plt
from numpy.polynomial.polynomial import polyfit
noise = np.random.uniform(-20, 20, 100)
x = np.linspace(0, 100, 100)+noise
y = np.linspace(0, 50, 100)+noise

plt.scatter(x, y, color='darkblue', edgecolor='red')
b, m = polyfit(x, y, 1)
y_pred = m*x+b
plt.plot(x,y_pred,"g",linewidth=2.5)
```

Download iris flower dataset from kaggle (https://www.kaggle.com/datasets/arshid/iris-flower-dataset) and read it in python dataframe

```python
#Add code here
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv("D:/5 semester/AI Lab/Lab11/IRIS.csv",
names=['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'class'])
print(data.head(15))
ir = data.head(50)
fig, axes = plt.subplots(figsize=(13,5))
axes.scatter(ir['sepal_length'], ir['sepal_width'])
axes.set_xlabel('sepal_length') 
axes.set_ylabel('sepal_width') 
axes.set_title("Iris Datast")

print(np.shape(data))
print(data.isnull().sum())
df = data.dropna()

print(data.duplicated().sum())
df = data.drop_duplicates()
print(np.shape(df))
axes.plot(ir['sepal_length'],ir['sepal_width'])
fig,axe = plt.subplots(1, 4, figsize=(15,5))
axe[0].set_title("bar plot")
axe[0].bar(ir['sepal_length'],ir['sepal_width'])
axe[1].set_title("step plot")
axe[1].step(ir['sepal_length'],ir['sepal_width'])
axe[2].set_title("fill_between plot")
axe[2].fill_between(ir['sepal_length'],ir['sepal_width'])
axe[3].set_title("plot")
axe[3].plot(ir['sepal_length'],ir['sepal_width'],color='green', marker='+', markersize=10, markerfacecolor='red')
axe[3].plot(ir['sepal_length'],ir['sepal_width'],color='blue', marker='s', markersize=10, markerfacecolor='blue')
axe[3].plot(ir['sepal_length'],ir['sepal_width'],color='green', marker='o', markersize=10, markerfacecolor='red')
```
